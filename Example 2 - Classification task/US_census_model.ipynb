### Imports, parameters and functions

```python
path='/home/roms/Kaggle/PayPlug/'

import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, roc_auc_score

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

def plotROC(real, predicted):
    fpr, tpr, thresholds = roc_curve(real, predicted, pos_label=1)
    plt.clf()
    plt.plot(fpr, tpr)
    plt.title('ROC curve')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.show()
```

### Data loading & retreatments
Headers have been created according to metadata info, instance weights are ignored, duplicates have been dropped, ' ?' values are considered as NA

```python
headers_corrected = pd.read_csv(path + 'Data/headers.csv',
                                delimiter='\t', index_col=1)
data_train = pd.read_csv(path + 'Data/census_income_learn.csv',
                         names=headers_corrected.index[:42], index_col=False, na_values=' ?')
data_train = data_train.drop('MARSUPWT', axis=1).drop_duplicates()
data_test = pd.read_csv(path + 'Data/census_income_test.csv',
                         names=headers_corrected.index[:42], index_col=False, na_values=' ?')
data_test = data_test.drop('MARSUPWT', axis=1)
data = pd.concat([data_train, data_test], axis=0, keys=['train', 'test'])
y = data['INCOME'].copy()
y[y == ' - 50000.'] = 0
y[y == ' 50000+.'] = 1
y = y.astype(int)
data = data.drop('INCOME', axis=1)
del data_train, data_test

headers_corrected
```

### Descriptive statistics
for nominal variables, computes basic statistics, along with top 10 value count and number of 'Not in universe' and NaN values<br>
for continuous ones, plots a histogram instead along with number of zeros and Nan values

```python
for column in data.columns:
    variable = headers_corrected.loc[column]['Description']
    variableType = headers_corrected.loc[column]['Type']
    series = data[column]['train']
    print('-'*40, '\n', 'VARIABLE: ', variable, '\n')
    if(variableType == 'nominal'):
        series = series.apply(str)
        print(series.describe(), '\n')
        if(series.describe()['unique'] <= 10):
            series.value_counts().plot(kind='bar', alpha=0.8)
        else:
            series.value_counts()[:10].plot(kind='bar', alpha=0.8)
        plt.title(variable.upper() + ' (counts)')
        plt.show()
        print('Not in universe count: ', (series == ' Not in universe')\
        .sum())
    else:
        print(series.describe(), '\n')
        series.hist(alpha=0.8, bins=20)
        plt.title(variable.upper() + ' (histogram)')
        plt.show()
        print('values = 0: ', (series == 0).sum())
    print('NaN values: ', pd.isnull(data[column]).sum(), '\n\n')

data = data.drop(['GRINST', 'MIGMTR1', 'MIGMTR3', 'MIGMTR4', 'MIGSUN',\
'PEFNTVTY', 'PEMNTVTY', 'PENATVTY'], axis=1)
```

When the value is not pertinent, 'Not in universe' is used. We also note that some value have outliers ('Wage per hour' for example). Some variables have lots of unknown values (everything related to migration for example).

### Features extraction
nominal variables are turned into binary variables ('Not in universe' is considered as a valid category)<br>
continuous variables are used as it<br>
a new variable 'Asset income' is calculated

```python
X = pd.DataFrame()
for column in data.columns:
    variableType = headers_corrected.loc[column]['Type']
    series = data[column]
    if(variableType == 'nominal'):
        dummies = pd.get_dummies(series)
        X[[column + '-' + str(x) for x in dummies.columns.tolist()]] = \
        dummies
    if(variableType == 'continuous'):
        X[column] = series

# gathers assets gains or losses into 1 var
X['ASSETSINC'] = X['CAPGAIN'] - X['CAPLOSS'] + X['DIVVAL']
X = X.drop(['CAPGAIN', 'CAPLOSS', 'DIVVAL'], axis=1)

del data
```

### Model training and validation
First, a Decision tree is trained on 80% of the training dataset, the rest is used for parameter 'max_depth' validation<br>
Then the model is trained again but using the most important features only<br>
Finally, a Random Forest model is trained in order to improve AUC score

#### Decision Trees

```python
# indexes for the training and validation sets
validationPart = 0.20
validIndex = np.random.choice(X.ix['train'].index, int(X.ix['train'].shape[0] \
                                * validationPart), replace=False)
trainIndex = list(set(X.ix['train'].index).difference(validIndex))
X = pd.concat([X.ix['train'].loc[trainIndex], X.ix['train'].loc[validIndex],
               X.ix['test']], axis=0, keys=['train', 'valid', 'test'])
y = pd.concat([y.ix['train'].loc[trainIndex], y.ix['train'].loc[validIndex],
               y.ix['test']], axis=0, keys=['train', 'valid', 'test'])
del trainIndex, validIndex
```

```python
# decision tree: we chose the parameter 'max_depth'
depths =  [2, 5, 10, 15, 20]
trainResult = []
validResult = []
for depth in depths:
    dt = DecisionTreeClassifier(max_depth=depth)
    dt.fit(X.ix['train'], y.ix['train'])
    trainResult.append(roc_auc_score(y.ix['train'], dt.predict_proba(X.ix['train'])[:,1]))
    validResult.append(roc_auc_score(y.ix['valid'], dt.predict_proba(X.ix['valid'])[:,1]))

plt.plot(depths, trainResult)
plt.plot(depths, validResult)
plt.xlabel('max_depth')
plt.ylabel('AUC')
plt.legend(['Training', 'Validation'], loc=0)
plt.title('AUC score = f(model complexity)')
plt.show()
# best: max_depth=10
```

From now on, we will use trees of max depth = 10.

```python
# trains the model again on both training and validation sets
class_weight={0:1, 1:3}
dt = DecisionTreeClassifier(max_depth=10, class_weight=class_weight)
dt.fit(X.ix[['train', 'valid']], y.ix[['train', 'valid']])
print('AUC score on test set: ', roc_auc_score(y.ix['test'], dt.predict_proba(X.ix['test'])[:,1]))
plotROC(y.ix['test'], dt.predict_proba(X.ix['test'])[:, 1])
export_graphviz(dt, out_file='tree.dot', max_depth=3, feature_names=X.columns)

# selects most important features only and trains the DT again on those features
importantFeatures = X.columns[dt.feature_importances_ > 0.01]
dt = DecisionTreeClassifier(max_depth=5, class_weight=class_weight)
dt.fit(X.ix[['train', 'valid']][importantFeatures], y.ix[['train', 'valid']])
print('\nAUC score on test set: ', roc_auc_score(y.ix['test'], dt.predict_proba(X.ix['test'][importantFeatures])[:,1]))
plotROC(y.ix['test'], dt.predict_proba(X.ix['test'][importantFeatures])[:, 1])
```

```python
print('Number of features: ', len(importantFeatures))
```

Using only 11 features, we have similar performances with the model using all the features.<br>
We have exported the decision tree in order to visualize the most important features. Unsurprisingly, those are: income from assets,  age, sex, number of weeks worked...

#### Random Forest

```python
# now we use a random forest model instead
class_weight={0:1, 1:3}
rf = RandomForestClassifier(max_depth=10, class_weight=class_weight)
rf.fit(X.ix[['train', 'valid']][importantFeatures], y.ix[['train', 'valid']])
print('AUC score on test set: ', roc_auc_score(y.ix['test'], rf.predict_proba(X.ix['test'][importantFeatures])[:,1]))
plotROC(y.ix['test'], rf.predict_proba(X.ix['test'][importantFeatures])[:, 1])
```
